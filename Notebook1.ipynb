Import CSV Data using read_csv

```python
import pandas as pd
df = pd.read_csv('jan2021.csv')
```

Show the columns available in the CSV

```python
df.columns
```

List how many of each type of Approvals there are

```python
df.groupby(['Submission Status']).count()
```

List the total approvals but sort them by types of approvals

```python
df.sort_values(['Submission Status'], ascending=False)
```

Search for all Submissions with the word containing "Orig". 

```python
import re
df.loc[df['Submission'].str.contains('orig[a-z]*', flags=re.I, regex=True)]
```

Group all Approvals by: Company, Submission Status, Submission, and Count. Then sort by ascending

```python
df['count'] = 1
df.groupby(['Company', 'Submission Status', 'Submission']).count()['count']
df
```

List # of approvals to each type of approval. Then append the Count to the table
