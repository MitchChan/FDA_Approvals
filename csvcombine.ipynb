```python
# Create master CSV file as master_df
# data from https://www.accessdata.fda.gov/scripts/cder/daf/index.cfm?event=reportsSearch.process
import pandas as pd
import os

master_df = pd.DataFrame()

for file in os.listdir(os.getcwd()):
    if file.endswith('.csv'):
        master_df = master_df.append(pd.read_csv(file))

master_df.to_csv('Master_File.csv', index=False)
```
